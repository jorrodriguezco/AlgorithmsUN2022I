#Dynamic programming

##Money Change Again

```python
import math

money = int(input())
denominations = [1, 3, 4]
minCoins = [0] + [math.inf]*money

for i in range(1, money+1):
    for j in denominations:
        if i>=j:
            coins = minCoins[i-j]+1
            if coins < minCoins[i]:
                minCoins[i] = coins

print(minCoins[money])
```

##Primitive Calculator

```python
import math

n = int(input())

# number of operations required for getting 0, 1, 2,.. , n
num_operations = [0, 0] + [math.inf]*(n-1)

for i in range(2, n+1):
    temp1, temp2, temp3 = [math.inf]*3

    temp1 = num_operations[i-1] + 1 
    if i%2 == 0: temp2 = num_operations[i//2] + 1
    if i%3 == 0: temp3 = num_operations[i//3] + 1
    min_ops = min(temp1, temp2, temp3)
    num_operations[i] = min_ops

print(num_operations[n])

# Backtracking the numbers leading to n
nums = [n]
while n!=1:
    if n%3 ==0 and num_operations[n]-1 == num_operations[n//3]:
        nums += [n//3]
        n = n//3
    elif n%2 ==0 and num_operations[n]-1 == num_operations[n//2]:
        nums += [n//2]
        n = n//2
    else:
        nums += [n-1]
        n = n - 1

print(' '.join([str(i) for i in nums][::-1]))
```

##Edit Distance

```python
def edit_distance(s, t):
    dis_s = len(s) + 1
    dis_t = len(t) + 1

    distance = [[x] + [0] * (dis_t - 1) for x in range(dis_s)]
    distance[0] = [x for x in range(dis_t)]

    for i in range(1, dis_s):
        for j in range(1, dis_t):

            if s[i - 1] == t[j - 1]:
                distance[i][j] = distance[i - 1][j - 1]
            else:
                distance[i][j] = min(
                    distance[i][j - 1], distance[i - 1][j], distance[i - 1][j - 1]) + 1

    return distance[-1][-1]


if __name__ == "__main__":
    print(edit_distance(input(), input()))
```

##Longest Common Subsequence of Two Sequences

```python
x = int(input())
a = list(map(int, input().split()))
y = int(input())
b = list(map(int, input().split()))

lcs = [[0 for i in range(y + 1)] for j in range(x + 1)]
for i in range(1, x+1):
    for j in range(1, y+1):
        if a[i-1] == b[j-1]:
            lcs[i][j] = lcs[i-1][j-1] + 1
        elif a[i-1] != b[j-1]:
            lcs[i][j] = max(lcs[i-1][j-1], lcs[i-1][j], lcs[i][j-1])
print(lcs[x][y])
```
