What is the size of the largest problem that you can solve in the time specified in row one?

## Time Complexity $n\lg(n)$

We will use the Lambert W function, which is defined as

$$f(x) = xe^x \Rightarrow f^{-1}(x) = W(x)$$

That done, we can calculate the inverse of $n\lg(n)$ as follows:

$$t = n\lg(n)$$

$$n = \frac{t\ln(2)}{W(t\ln(2))}$$

### Second:

$$
\begin{align*}
  n\lg(n) \times 10^{-9} \text{ s} &\leq 1 \text{ s} \\
  n\lg(n) &\leq 10^9 \\
  n &\leq 3.962008 \times 10^7
\end{align*}
$$

```python
from scipy.special import *

def inverse_nlgn(t) -> float:
  solution = (t*math.log(2)) / lambertw(t*math.log(2)).real
  return "{:e}".format(solution)
```

```python
print("Second:    ", inverse_nlgn(10**9))
print("Minute:    ", inverse_nlgn(6*(10**10)))
print("Hour:      ", inverse_nlgn(3.6*(10**12)))
print("Day:       ", inverse_nlgn(8.64*(10**13)))
print("Year:      ", inverse_nlgn(3.1536*(10**16)))
print("Century:   ", inverse_nlgn(3.1536*(10**18)))
print("Millennium:", inverse_nlgn(3.1536*(10**19)))
```

```python
def fact(n):
  return math.factorial(n)

def findFact(n):
  i = 0
  while fact(i) < n:
    i += 1
  
  return i-1
```

```python
print("Second:    ", findFact(10**9))
print("Minute:    ", findFact(6*(10**10)))
print("Hour:      ", findFact(3.6*(10**12)))
print("Day:       ", findFact(8.64*(10**13)))
print("Year:      ", findFact(3.1536*(10**16)))
print("Century:   ", findFact(3.1536*(10**18)))
print("Millennium:", findFact(3.1536*(10**19)))
```

  n^n \times 10^{-9} \text{ s} &\leq 3.1536 \times 10^7 \text{ s} \\
  n^n &\leq 3.1536 \times 10^{16} \\
  n &\leq 14
\end{align*}
$$

### Century:

$$
\begin{align*}
  n^n \times 10^{-9} \text{ s} &\leq 3.1536 \times 10^9 \text{ s} \\
  n^n &\leq 3.1536 \times 10^{18} \\
  n &\leq 15
\end{align*}
$$

### Millennium:

$$
\begin{align*}
  n^n \times 10^{-9} \text{ s} &\leq 3.1536 \times 10^{10} \text{ s} \\
  n^n &\leq 3.1536 \times 10^{19} \\
  n &\leq 16
\end{align*}
$$

```python
def findNN(n):
  i = 0
  while i**i < n:
    i += 1
  
  return i-1
```

```python
print("Second:    ", findNN(10**9))
print("Minute:    ", findNN(6*(10**10)))
print("Hour:      ", findNN(3.6*(10**12)))
print("Day:       ", findNN(8.64*(10**13)))
print("Year:      ", findNN(3.1536*(10**16)))
print("Century:   ", findNN(3.1536*(10**18)))
print("Millennium:", findNN(3.1536*(10**19)))
```
